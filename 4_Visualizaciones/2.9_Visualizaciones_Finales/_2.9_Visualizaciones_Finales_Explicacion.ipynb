## 1. Distribución de transacciones por categoría de producto  
**Archivo generado:** `1_final_distribucion_categorias.png`

### Descripción  
Gráfico de barras que muestra el número total de transacciones por categoría de producto:  
- Clothing  
- Electronics  
- Beauty  

Permite observar de forma directa qué categorías concentran mayor actividad de compra dentro del periodo analizado.

### Relación con lo visto en clase  
En los notebooks de EDA se utilizó `value_counts()` y gráficos de barras para analizar variables categóricas.  
Esta visualización aplica exactamente la misma técnica, pero con un enfoque de cierre para el proyecto, pensando ya en el reporte ejecutivo.

### Utilidad analítica  
Ayuda a identificar:

- Categorías con mayor volumen de transacciones  
- Posibles categorías rezagadas  
- Equilibrio o desequilibrio en la distribución del portafolio  

### Relevancia empresarial  
Es clave para:

- Toma de decisiones sobre inventario y surtido  
- Diseño de promociones específicas por categoría  
- Priorización de inversiones en líneas de producto estratégicas

## 2. Ventas totales por mes  
**Archivo generado:** `2_final_ventas_totales_mes.png`

### Descripción  
Gráfico de líneas que muestra el monto total vendido en cada mes del año.  
Permite identificar picos y valles a lo largo del tiempo, evidenciando estacionalidad en las ventas.

### Relación con lo visto en clase  
Durante las prácticas de transformación se generaron columnas derivadas de fecha (como mes) utilizando `.dt.month`.  
Esta visualización utiliza esa transformación para representar el comportamiento temporal de las ventas, tal como se hizo en ejercicios previos de series de tiempo sencillas.

### Utilidad analítica  
Permite:

- Detectar meses con mayor y menor demanda  
- Observar posibles patrones cíclicos  
- Evaluar el impacto de temporadas altas o bajas  

### Relevancia empresarial  
Sirve para:

- Planear campañas de marketing estacional  
- Ajustar inventarios y logística según el calendario  
- Estimar de manera más informada los ingresos mensuales esperados

## 3. Distribución de la edad de los clientes  
**Archivo generado:** `3_final_distribucion_edad.png`

### Descripción  
Histograma que muestra la distribución de la edad de los clientes en el dataset.  
Se observa cómo se concentran las frecuencias en distintos rangos de edad.

### Relación con lo visto en clase  
En el EDA de otros datasets (como viviendas y prácticas de normalización) se utilizaron histogramas para entender la forma de las variables numéricas y su dispersión.  
Aquí se aplica el mismo enfoque a la variable demográfica clave: la edad.

### Utilidad analítica  
Permite:

- Conocer qué grupos de edad predominan en la base de clientes  
- Detectar si la población es homogénea o muy variada  
- Servir como base para segmentaciones posteriores (por ejemplo, rangos de edad)  

### Relevancia empresarial  
Es importante para:

- Diseñar campañas dirigidas a segmentos etarios específicos  
- Ajustar el tipo de producto o mensaje según el perfil del cliente  
- Identificar oportunidades en rangos de edad poco atendidos

## 4. Matriz de correlación entre variables numéricas  
**Archivo generado:** `4_final_matriz_correlacion.png`

### Descripción  
Mapa de calor que muestra las correlaciones entre las variables numéricas principales:

- edad  
- cantidad  
- precio_unitario  
- monto_total  

Los valores cercanos a 1 indican correlaciones fuertes y positivas; valores cercanos a 0 indican poca relación lineal.

### Relación con lo visto en clase  
En EDA y en los modelos de regresión se utilizaron heatmaps de correlación para seleccionar variables relevantes y entender la relación entre ellas (por ejemplo, en BostonHousing y en el propio Retail EDA).  
Esta matriz final resume esas relaciones en el contexto del proyecto.

### Utilidad analítica  
Permite:

- Confirmar que `monto_total` está fuertemente relacionado con `precio_unitario` y `cantidad`  
- Detectar posibles colinealidades entre variables  
- Justificar la selección de atributos en los modelos de Machine Learning  

### Relevancia empresarial  
Ayuda a:

- Identificar los factores que más influyen en el monto de compra  
- Entender cómo interactúan el precio y la cantidad en el resultado final  
- Priorizar variables clave en análisis futuros o simulaciones de escenarios
