## 1. Matriz de confusión  
**Archivo generado:** `1_log_matriz_confusion.png`

### Descripción  
La matriz de confusión permite evaluar el desempeño del modelo de regresión logística al clasificar clientes como:

- 1 = Cliente de alto gasto (High Spender)  
- 0 = Cliente de gasto bajo o moderado  

La visualización muestra:

- Verdaderos negativos (150)  
- Falsos positivos (0)  
- Falsos negativos (7)  
- Verdaderos positivos (143)  

### Relación con lo visto en clase  
Durante la práctica **RL_01.ipynb (BankNote Authentication)** se empleó la matriz de confusión para medir el rendimiento de modelos de clasificación.  
Aquí se replica esa misma metodología aplicada al contexto de ventas retail.  

El objetivo es identificar la capacidad del modelo para clasificar correctamente a los clientes según su nivel de gasto.

### Utilidad analítica  
La matriz permite observar:

- Si el modelo comete más errores al predecir una clase específica.  
- Qué tan bien distingue entre clientes de bajo y alto gasto.  
- El equilibrio entre sensibilidad (recall) y precisión.  

En este caso, la cantidad de errores es baja, predominando las predicciones correctas.

### Relevancia empresarial  
Una matriz de confusión precisa respalda decisiones como:

- Identificación de clientes valiosos.  
- Diseñar promociones según segmentos.  
- Priorizar campañas para clientes con mayor probabilidad de compra.  

El buen desempeño del modelo indica que puede emplearse como herramienta de segmentación.

## 2. Curva ROC y AUC  
**Archivo generado:** `2_log_curva_roc.png`

### Descripción  
La curva ROC muestra la relación entre:

- **Tasa de verdaderos positivos (TPR)**  
- **Tasa de falsos positivos (FPR)**  

El área bajo la curva (AUC = 0.988) cuantifica la capacidad del modelo para distinguir entre ambas clases.

Un AUC cercano a 1 indica un desempeño excelente.

### Relación con lo visto en clase  
En la práctica de regresión logística del archivo **RL_01.ipynb**, se generó la curva ROC para evaluar modelos binarios.  
Este proyecto adopta la misma técnica aplicada al Retail Sales Dataset, garantizando consistencia metodológica con lo aprendido en clase.

### Utilidad analítica  
La curva ROC permite:

- Evaluar el rendimiento del modelo independientemente del umbral.  
- Comparar modelos de clasificación.  
- Determinar qué tan bien separa las clases 0 y 1.  

El AUC de 0.988 indica que el modelo tiene una capacidad casi perfecta para diferenciar clientes de alto gasto.

### Relevancia empresarial  
Una curva ROC con alto AUC es fundamental para tareas como:

- Selección de clientes potenciales.  
- Pronóstico de comportamiento de compra.  
- Estrategias de marketing y retención.  

El modelo puede utilizarse para priorizar clientes con mayor probabilidad de efectuar compras importantes.

Las visualizaciones generadas demuestran que el modelo de regresión logística presenta un excelente rendimiento tanto en métricas de clasificación como en separación entre clases.  
Tanto la matriz de confusión como la curva ROC confirman su utilidad para segmentar clientes y apoyar decisiones estratégicas en el área comercial.
