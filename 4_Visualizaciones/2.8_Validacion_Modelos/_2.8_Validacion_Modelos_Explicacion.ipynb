## 1. Matriz de confusión – Validación  
**Archivo generado:** `1_validacion_matriz_confusion.png`

### Descripción  
La matriz de confusión representa los resultados de la validación del modelo de Regresión Logística mediante los valores reales y las predicciones generadas.  
Incluye los cuatro componentes principales:

- Verdaderos negativos  
- Verdaderos positivos  
- Falsos positivos  
- Falsos negativos  

Esta visualización permite confirmar si el modelo mantiene un comportamiento consistente respecto al conjunto de prueba.

### Relación con lo visto en clase  
En la práctica **RL_01.ipynb (BankNote)** la matriz de confusión fue utilizada como herramienta principal de validación para modelos supervisados.  
Este proyecto replica exactamente esa metodología, aplicándola al análisis de clientes retail.

### Utilidad analítica  
La matriz evidencia la estabilidad del modelo al ser evaluado nuevamente en un proceso aislado.  
Permite identificar:

- La capacidad de generalización del modelo  
- El desempeño fuera del entrenamiento  
- Posibles desequilibrios en la clasificación  

### Relevancia empresarial  
Garantiza que el modelo no solo funciona en el entrenamiento, sino también con nuevos datos.  
Esto es esencial para:

- Segmentación de clientes  
- Campañas basadas en probabilidad de compra  
- Decisiones de marketing apoyadas en evidencia estadística

## 2. Curva ROC y AUC – Validación  
**Archivo generado:** `2_validacion_curva_roc.png`

### Descripción  
La curva ROC muestra el comportamiento del modelo validado en términos de sensibilidad y tasa de falsos positivos.  
El área bajo la curva (AUC) actúa como un indicador global de rendimiento.

### Relación con lo visto en clase  
En la práctica de regresión logística, la curva ROC se utilizó para evaluar la capacidad discriminativa de modelos binarios.  
En esta validación se reproduce la misma técnica con datos reales del proyecto, garantizando coherencia metodológica.

### Utilidad analítica  
El valor AUC muestra si el modelo mantiene un desempeño consistente durante el proceso de validación.  
Una curva elevada indica:

- Buena separabilidad entre clases  
- Interpretación clara del desempeño  
- Robustez del modelo ante nuevos datos  

### Relevancia empresarial  
Un modelo con alto AUC es adecuado para:

- Identificar clientes potenciales  
- Clasificar perfiles de compra  
- Priorizar recursos hacia segmentos valiosos  

La estabilidad del AUC en la validación indica que el modelo puede integrarse en análisis empresariales sin riesgos de sobreajuste severo.

---

## Conclusión General  
La validación confirma que el modelo conserva un buen desempeño al ser evaluado con datos nuevos.  
Tanto la matriz de confusión como la curva ROC muestran un comportamiento estable, consistente y adecuado para su uso en procesos de segmentación y toma de decisiones empresariales.
