## 1. Valores reales vs. valores predichos  
**Archivo generado:** `1_reg_lineal_real_vs_predicho.png`

### Descripción  
Esta visualización compara los valores reales del monto total estandarizado (`monto_total_std`) con los valores predichos por el modelo de regresión lineal.  
Cada punto corresponde a una observación del conjunto de prueba.  
La línea diagonal roja representa el comportamiento ideal donde las predicciones coinciden con los valores reales.

### Relación con lo visto en clase  
La gráfica sigue la misma metodología aplicada en los notebooks de regresión lineal revisados en clase:  
- BostonHousing.ipynb  
- MLineral2.ipynb  

En esas prácticas se utilizó un scatterplot para evaluar visualmente la calidad del ajuste del modelo. El presente proyecto replica ese procedimiento aplicado al Retail Sales Dataset.

### Utilidad analítica  
La gráfica permite identificar:  
- El grado de ajuste del modelo.  
- La cercanía de las predicciones con los valores reales.  
- La presencia de errores sistemáticos o dispersión anómala.  

El comportamiento ascendente confirma que el modelo aprendió una relación lineal entre las variables predictoras y el monto total.

### Relevancia empresarial  
Una predicción fiable del monto total gastado por un cliente puede utilizarse para:  
- Realizar proyecciones de ventas.  
- Analizar el comportamiento de gasto por segmentos de clientes.  
- Identificar clientes de alto valor.  
- Optimizar estrategias de marketing y promociones.

### Conclusión  
La visualización demuestra que el modelo presenta un comportamiento coherente, con predicciones razonablemente alineadas con los valores reales.  
Si bien existe dispersión, la tendencia general valida que el modelo puede emplearse como herramienta de análisis exploratorio y apoyo para decisiones comerciales.
